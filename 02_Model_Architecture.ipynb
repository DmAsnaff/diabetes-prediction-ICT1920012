```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Here datset is loaded and printing the datashape and the outcome distribution (with normalized)
df = pd.read_csv('diabetes.csv')
print(f"Initial data shape: {df.shape}")
print(f"Class distribution:\n{df.Outcome.value_counts(normalize=True)}")


# Handled impossible zeros and replace with NaN (medical measurements can't be 0)
features = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[features] = df[features].replace(0, np.nan)

# Imputeing missing values with median because some ML algo cannot handle NaN values directly
imputer = SimpleImputer(strategy='median')
df[features] = imputer.fit_transform(df[features])

# Normalizationing to adjusts the range and distribution of features
scaler = StandardScaler()
X = scaler.fit_transform(df.drop('Outcome', axis=1))
y = df.Outcome.values


# Stratified splitting to maintain class ratio
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=42  # 0.15/0.85 â‰ˆ 0.1765
)
print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")


```

# Baseline Neural Network Architecture (No Regularization)

```python
from tensorflow.keras import models, layers

baseline_model = models.Sequential([
    layers.Dense(16, activation='relu', input_shape=(8,)),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC, Precision, Recall

baseline_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', AUC(name='auc'), Precision(name='precision'), Recall(name='recall')]
)
```

```python
history = baseline_model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    verbose=1
)
```

```python
import matplotlib.pyplot as plt

def plot_history(history, model_name='baseline'):
    # Plot Loss
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title(f'{model_name}: Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title(f'{model_name}: Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(f'{model_name}_training_curves_withoutany.png')
    plt.show()
    
plot_history(history, model_name='baseline')

```

```python
test_metrics = baseline_model.evaluate(X_test, y_test, verbose=0)

print(f"Test Accuracy:  {test_metrics[1]:.4f}")
print(f"Test AUC:       {test_metrics[2]:.4f}")
print(f"Test Precision: {test_metrics[3]:.4f}")
print(f"Test Recall:    {test_metrics[4]:.4f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predict class labels
y_pred_prob = baseline_model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Diabetic", "Diabetic"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix on Test Set")
plt.savefig("baseline_Withoutanymodel_confusion_matrix.png")
plt.show()
```

```python
train_metrics = baseline_model.evaluate(X_train, y_train, verbose=0)
val_metrics = baseline_model.evaluate(X_val, y_val, verbose=0)

print(f"Train Accuracy:  {train_metrics[1]:.4f}")
print(f"Train AUC:       {train_metrics[2]:.4f}")
print(f"Train Precision: {train_metrics[3]:.4f}")
print(f"Train Recall:    {train_metrics[4]:.4f}")

print(f"Val Accuracy:    {val_metrics[1]:.4f}")
print(f"Val AUC:         {val_metrics[2]:.4f}")
print(f"Val Precision:   {val_metrics[3]:.4f}")
print(f"Val Recall:      {val_metrics[4]:.4f}")
```

# Regularized Model Architecture

```python
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping

regularized_model = models.Sequential([
    layers.Dense(16, kernel_regularizer=regularizers.l2(0.01), input_shape=(8,)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.2),

    layers.Dense(16, kernel_regularizer=regularizers.l2(0.01)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.2),

    layers.Dense(1, activation='sigmoid')
])
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import AUC, Precision, Recall


regularized_model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', AUC(name='auc'), Precision(name='precision'), Recall(name='recall')]
)

early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history_reg = regularized_model.fit(
    X_train, y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_val, y_val),
    callbacks=[early_stop],
    verbose=1
)
```

```python
def plot_history(history, model_name='model'):
    import matplotlib.pyplot as plt
    import numpy as np

    # Extract values
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']
    train_acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    
    # Identify best epoch (lowest val_loss)
    best_epoch = np.argmin(val_loss)

    plt.figure(figsize=(12, 5))

    # ðŸ“‰ Loss Plot
    plt.subplot(1, 2, 1)
    plt.plot(train_loss, label='Train Loss')
    plt.plot(val_loss, label='Val Loss')
    plt.axvline(best_epoch, color='red', linestyle='--', label=f'Stopped @ {best_epoch}')
    plt.title(f'{model_name}: Loss over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # ðŸ“ˆ Accuracy Plot
    plt.subplot(1, 2, 2)
    plt.plot(train_acc, label='Train Accuracy')
    plt.plot(val_acc, label='Val Accuracy')
    plt.axvline(best_epoch, color='red', linestyle='--', label=f'Stopped @ {best_epoch}')
    plt.title(f'{model_name}: Accuracy over Epochs')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(f'{model_name}_training_curves.png', dpi=300)
    plt.show()
plot_history(history_reg, model_name='regularized')
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predict class labels from test probabilities
y_pred_prob = regularized_model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Non-Diabetic", "Diabetic"])
disp.plot(cmap='Blues')
plt.title("Regularized Model: Confusion Matrix (Test Set)")
plt.savefig("regularized_confusion_matrix.png", dpi=300)
plt.show()
```

```python
# Evaluate on all three sets
train_metrics = regularized_model.evaluate(X_train, y_train, verbose=0)
val_metrics = regularized_model.evaluate(X_val, y_val, verbose=0)
test_metrics = regularized_model.evaluate(X_test, y_test, verbose=0)

def report_metrics(metrics, label=''):
    print(f"\n{label} Set:")
    print(f"  Accuracy : {metrics[1]:.4f}")
    print(f"  AUC      : {metrics[2]:.4f}")
    print(f"  Precision: {metrics[3]:.4f}")
    print(f"  Recall   : {metrics[4]:.4f}")

report_metrics(train_metrics, label='Train')
report_metrics(val_metrics, label='Validation')
report_metrics(test_metrics, label='Test')
```

```python
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.metrics import AUC, Precision, Recall

#Train each using the same way 

# Common callback
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Common compile args
compile_args = {
    'optimizer': Adam(learning_rate=0.001),
    'loss': 'binary_crossentropy',
    'metrics': ['accuracy', AUC(name='auc'), Precision(name='precision'), Recall(name='recall')]
}


model_a = models.Sequential([
    layers.Dense(16, kernel_regularizer=regularizers.l2(0.01), input_shape=(8,)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')
])

model_a.compile(**compile_args)

history_a = model_a.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix
import numpy as np

def evaluate_and_log(model, name, X_train, y_train, X_val, y_val):
    print(f"\n========== ðŸ“Š {name} ==========")

    # Predict probabilities
    y_train_prob = model.predict(X_train)
    y_val_prob = model.predict(X_val)

    # Class predictions
    y_train_pred = (y_train_prob > 0.5).astype(int)
    y_val_pred = (y_val_prob > 0.5).astype(int)

    # Compute metrics
    train_acc = np.mean(y_train_pred.flatten() == y_train)
    val_acc = np.mean(y_val_pred.flatten() == y_val)
    train_auc = roc_auc_score(y_train, y_train_prob)
    val_auc = roc_auc_score(y_val, y_val_prob)

    print(f"Train Accuracy : {train_acc:.4f}")
    print(f"Val Accuracy   : {val_acc:.4f}")
    print(f"Train AUC      : {train_auc:.4f}")
    print(f"Val AUC        : {val_auc:.4f}")

    print("\nValidation Classification Report:")
    print(classification_report(y_val, y_val_pred, digits=4))

    # Confusion matrix
    cm = confusion_matrix(y_val, y_val_pred)
    print("Confusion Matrix (Validation):")
    print(cm)

    return {
        'name': name,
        'train_acc': train_acc,
        'val_acc': val_acc,
        'train_auc': train_auc,
        'val_auc': val_auc,
        'confusion_matrix': cm
    }
results_a = evaluate_and_log(model_a, "Variant A (1 Hidden Layer)", X_train, y_train, X_val, y_val)
```
