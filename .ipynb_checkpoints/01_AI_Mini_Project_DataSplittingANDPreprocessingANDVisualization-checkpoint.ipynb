```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Here datset is loaded and printing the datashape and the outcome distribution (with normalized)
df = pd.read_csv('diabetes.csv')
print(f"Initial data shape: {df.shape}")
print(f"Class distribution:\n{df.Outcome.value_counts(normalize=True)}")
```

```python

# Handled impossible zeros and replace with NaN (medical measurements can't be 0)
features = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[features] = df[features].replace(0, np.nan)
```

```python
# Imputeing missing values with median because some ML algo cannot handle NaN values directly
imputer = SimpleImputer(strategy='median')
df[features] = imputer.fit_transform(df[features])

```

```python
# Visualize distributions of each feature 
plt.figure(figsize=(12, 8))
for i, col in enumerate(df.columns[:-1], 1):
    plt.subplot(3, 3, i)
    sns.histplot(data=df, x=col, kde=True, hue='Outcome', multiple='stack')
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.savefig('feature_distributions.png')
```

```python

# Normalizationing to adjusts the range and distribution of features
scaler = StandardScaler()
X = scaler.fit_transform(df.drop('Outcome', axis=1))
y = df.Outcome.values
```

```python

# Stratified splitting to maintain class ratio
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=42  # 0.15/0.85  0.1765
)
print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")
```

```python
# To show the class distribution after split

# here create DataFrames for each split
train_set = pd.DataFrame(X_train, columns=df.columns[:-1])
train_set['Outcome'] = y_train
train_set['Set'] = 'Train'

val_set = pd.DataFrame(X_val, columns=df.columns[:-1])
val_set['Outcome'] = y_val
val_set['Set'] = 'Validation'

test_set = pd.DataFrame(X_test, columns=df.columns[:-1])
test_set['Outcome'] = y_test
test_set['Set'] = 'Test'

# Combineing for visualization
combined_set = pd.concat([train_set, val_set, test_set])

# Plot class distribution per set
plt.figure(figsize=(10, 6))
sns.countplot(x='Set', hue='Outcome', data=combined_set, palette='viridis')
plt.title('Class Distribution Across Splits (Stratified)')
plt.ylabel('Count')
plt.savefig('split_class_distribution.png', bbox_inches='tight')
```
