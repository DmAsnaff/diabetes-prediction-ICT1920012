```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Here datset is loaded and printing the datashape and the outcome distribution (with normalized)
df = pd.read_csv('diabetes.csv')
print(f"Initial data shape: {df.shape}")
print(f"Class distribution:\n{df.Outcome.value_counts(normalize=True)}")


# Handled impossible zeros and replace with NaN (medical measurements can't be 0)
features = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[features] = df[features].replace(0, np.nan)

# Imputeing missing values with median because some ML algo cannot handle NaN values directly
imputer = SimpleImputer(strategy='median')
df[features] = imputer.fit_transform(df[features])

# Normalizationing to adjusts the range and distribution of features
scaler = StandardScaler()
X = scaler.fit_transform(df.drop('Outcome', axis=1))
y = df.Outcome.values


# Stratified splitting to maintain class ratio
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.1765, stratify=y_temp, random_state=42  # 0.15/0.85 â‰ˆ 0.1765
)
print(f"Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")


```

```python
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.metrics import AUC, Precision, Recall


def build_model(units=16, dropout_rate=0.2, l2_lambda=0.01, learning_rate=0.001):
    model = models.Sequential([
        layers.Dense(units, kernel_regularizer=regularizers.l2(l2_lambda), input_shape=(8,)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Dropout(dropout_rate),
        
        layers.Dense(units, kernel_regularizer=regularizers.l2(l2_lambda)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Dropout(dropout_rate),

        layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss='binary_crossentropy',
        metrics=['accuracy', AUC(name='auc'), Precision(name='precision'), Recall(name='recall')]
    )
    
    return model
```

```python
import numpy as np
from itertools import product

from sklearn.metrics import roc_auc_score

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import models, layers, regularizers
from tensorflow.keras.optimizers import Adam

param_grid = {
    'units': [8, 16, 32],
    'dropout': [0.1, 0.2, 0.3],
    'l2': [0.001, 0.01, 0.1],
    'lr': [0.0005, 0.001, 0.005],
    'batch': [16, 32, 64]
}

top_results = []
combo_count = 1

for units, dropout, l2_val, lr, batch_size in product(
        param_grid['units'], param_grid['dropout'], param_grid['l2'], param_grid['lr'], param_grid['batch']):
    
    print(f"\nðŸš€ Training Combo {combo_count}: units={units}, dropout={dropout}, l2={l2_val}, lr={lr}, batch={batch_size}")
    model = build_model(units, dropout, l2_val, lr)
    
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    
    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=100,
        batch_size=batch_size,
        callbacks=[early_stop],
        verbose=0
    )

    val_auc = roc_auc_score(y_val, model.predict(X_val))
    val_acc = model.evaluate(X_val, y_val, verbose=0)[1]

    top_results.append((val_auc, val_acc, units, dropout, l2_val, lr, batch_size))
    combo_count += 1

# Sort by AUC, then Accuracy
top_results = sorted(top_results, key=lambda x: (x[0], x[1]), reverse=True)
```

```python
import pandas as pd

results_df = pd.DataFrame(top_results, columns=[
    'val_auc', 'val_accuracy', 'units', 'dropout', 'l2', 'learning_rate', 'batch_size'
])

# Sort by AUC and then accuracy
results_df = results_df.sort_values(by=['val_auc', 'val_accuracy'], ascending=False)
print(results_df.head(10))
```

```python
best_config = results_df.iloc[0]
print("\n Best Configuration Found:")
print(best_config)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid', font_scale=1.1)

def plot_metric_vs_param(df, param, metric='val_auc'):
    plt.figure(figsize=(7, 5))
    sns.boxplot(x=param, y=metric, data=df)
    plt.title(f'{metric.upper()} vs {param}')
    plt.xlabel(param)
    plt.ylabel(metric)
    plt.tight_layout()
    plt.show()

# Plot against each hyperparameter
plot_metric_vs_param(results_df, 'units', 'val_auc')
plot_metric_vs_param(results_df, 'dropout', 'val_auc')
plot_metric_vs_param(results_df, 'l2', 'val_auc')
plot_metric_vs_param(results_df, 'learning_rate', 'val_auc')
plot_metric_vs_param(results_df, 'batch_size', 'val_auc')

# You can also do accuracy plots:
plot_metric_vs_param(results_df, 'units', 'val_accuracy')
```
